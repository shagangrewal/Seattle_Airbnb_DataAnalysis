```python
#Applying Linear Regression model to the dataset
#Before applying the model, we need to figure the important columns for the dataset
#Since we have some outliers in our dataset, we will refine it to have better dataset to train and test
price_df = listings_subset.loc[listings_subset['price']<425]
```

```python
#creating numerical values for categorical values in the dataset
price_df['room_type'].replace(to_replace=['Entire home/apt','Private room','Shared room'], value=['1','2','3'], inplace=True)
price_df['room_type'] = price_df['room_type'].astype(str).astype(int)
```

```python
price_df['host_is_superhost'].replace(to_replace=['t','f'], value=['1','0'],inplace=True)
price_df['host_is_superhost'] = price_df['host_is_superhost'].astype(str).astype(float)
```

```python
price_df.dtypes
```

```python
#dropping the categorical columns
price_df = price_df.drop(['host_since','neighbourhood_cleansed','property_type' ,'amenities','cancellation_policy'],axis=1)
```

```python
price_df.dtypes
```

```python
#Linear Regression
#Diving into explanatory and responsive variables

X = price_df.drop(['price','id','host_id'],axis=1)
Y = price_df['price']
```

```python
#To take care of the NaN and infinite values

X = np.nan_to_num(X)
Y = np.nan_to_num(Y)
```

```python
#splitting data into train and testing 
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
#applying Linear Regression model and then fitting the data
from sklearn.preprocessing import StandardScaler

LM_model = make_pipeline(StandardScaler(with_mean=False), LinearRegression()) 
LM_model.fit(x_train, y_train)
```

```python
ytest_pred = LM_model.predict(x_test)
ytrain_pred = LM_model.predict(x_train)
```

```python
#evaluating the training and testing score

train_score = r2_score(y_test,ytest_pred)
test_score = r2_score(y_train, ytrain_pred)
```

```python
print('The rsquared on training data: {0}, while the rsquared on test data: {1}'.format(train_score, test_score))
```
