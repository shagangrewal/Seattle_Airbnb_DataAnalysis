```python
#as visible from the heatmap above, a lot of variables are not correlated
#below code is used to take less correlated variables and then eliminate the ones which are not useful for us
#this can be referred to as feature selection as only the features important to the data will be dropped

not_to_drop = [ 'id','host_id', 'host_since', 'host_is_superhost', 'host_listings_count', 'host_total_listings_count', 'neighbourhood_cleansed', 'property_type', 'room_type',
       'latitude', 'longitude', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'amenities','price', 'guests_included', 'minimum_nights', 'maximum_nights', 'availability_30', 'availability_60', 'availability_90', 'availability_365', 'number_of_reviews', 'cancellation_policy', 'review_scores_value', 'calculated_host_listings_count',
       'reviews_per_month']

to_drop = [col for col in df_listings.columns if col not in not_to_drop]

listings_subset = df_listings.drop(to_drop,axis=1)

listings_subset.shape
```

```python
#Cleaning and transformation of data
#price is object type(categorical data) converting it to float

object_to_float = lambda col: col.replace('$','').replace(',','')
listings_subset['price'] = listings_subset['price'].apply(object_to_float).astype(float)
```
